## Setup

Este notebook espera um arquivo `.env` na raiz do projeto com `OPENAI_API_KEY`.

```python
import os
import json
from pathlib import Path

import requests
from dotenv import load_dotenv

env_path = Path.cwd() / ".env"
load_dotenv(env_path)

api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("API key nao encontrada")

headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json",
}
```

## API antiga: `chat/completions`

```python
url = "https://api.openai.com/v1/chat/completions"

data = {
    "model": "gpt-3.5-turbo",
    "messages": [
        {"role": "system", "content": "Voce e um especialista em Inteligencia Artificial Generativa."},
        {"role": "user", "content": "Explique o que sao LLMs e como ganhar dinheiro com eles."},
    ],
    "max_tokens": 500,
    "temperature": 0.7,
}

response = requests.post(url, headers=headers, data=json.dumps(data))
```

```python
print("=== Status da Resposta ===")
print(f"Status Code       : {response.status_code}")
print(f"Motivo            : {response.reason}")
print(f"Sucesso?          : {response.ok}")

print("\n=== Cabecalhos da Resposta ===")
for key, value in response.headers.items():
    print(f"{key}: {value}")

print("\n=== Corpo da Resposta ===")
print(f"Texto (raw)       : {response.text[:200]}...")
print(f"JSON (parsed)     : {response.json()}")

print("\n=== Informacoes da Solicitacao (Request) ===")
print(f"Metodo HTTP       : {response.request.method}")
print(f"URL               : {response.request.url}")
print(f"Cabecalhos Enviados: {response.request.headers}")
print(f"Payload Enviado   : {response.request.body}")

print("\n=== Tempo de Resposta ===")
print(f"Tempo de resposta : {response.elapsed}")

print("\n=== Cookies (se houver) ===")
print(f"Cookies recebidos : {response.cookies}")

print("\n=== Historico de Redirecionamento ===")
print(f"Historico de Redirects: {response.history}")
```

```python
resposta = response.json()
print(resposta["choices"][0]["message"]["content"])
```

## API nova: `responses`

```python
url = "https://api.openai.com/v1/responses"

data = {
    "model": "gpt-3.5-turbo",
    "input": "Voce e um especialista em Inteligencia Artificial Generativa.\nExplique o que sao LLMs e como ganhar dinheiro com eles.",
}

response = requests.post(url, headers=headers, json=data)
resposta = response.json()
print(resposta["output"][0]["content"][0]["text"])
```
