5. OPERAÇÕES COM PONTOS
   Crie um conjunto de funções para trabalhar com pontos no plano cartesiano,
   onde cada ponto é representado por uma tupla (x, y):
   
   - `distancia(p1, p2)`: calcula a distância entre dois pontos
   - `ponto_medio(p1, p2)`: retorna o ponto médio como uma tupla
   - `inclinacao(p1, p2)`: calcula a inclinação da reta (se possível)
   - `quadrante(p)`: retorna o quadrante do ponto (1, 2, 3, 4 ou 0 se estiver sobre um eixo)

```python
def distancia(p1, p2):
    """Calcula a distância entre dois pontos p1 e p2."""
    x1, y1 = p1
    x2, y2 = p2
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
def ponto_medio(p1, p2):
    """Retorna o ponto médio entre p1 e p2."""
    x1, y1 = p1
    x2, y2 = p2
    return ((x1 + x2) / 2, (y1 + y2) / 2)
def inclinacao(p1, p2):
    """Calcula a inclinação da reta entre p1 e p2, se possível."""
    x1, y1 = p1
    x2, y2 = p2
    if x2 - x1 == 0:
        return None  # Reta vertical, inclinação indefinida
    return (y2 - y1) / (x2 - x1)
def quadrante(p):
    """Retorna o quadrante do ponto p."""
    x, y = p
    if x > 0 and y > 0:
        return 1
    elif x < 0 and y > 0:
        return 2
    elif x < 0 and y < 0:
        return 3
    elif x > 0 and y < 0:
        return 4
    else:
        return 0  # Ponto sobre um eixo
```

```python
# Exemplo de uso:
p1 = (2, 3)
p2 = (5, 7)
print("Distância:", distancia(p1, p2))
print("Ponto médio:", ponto_medio(p1, p2))
print("Inclinação:", inclinacao(p1, p2))
print("Quadrante de p1:", quadrante(p1))
print("Quadrante de p2:", quadrante(p2))
```
